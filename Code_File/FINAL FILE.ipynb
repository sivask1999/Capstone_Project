# Statistical test

```python
cont_var=["iralcfy",'irmjfy', 'ircocfy', 'irhallucyfq', 'irinhalyfq','wrkdhrswk2','irherfy','irmethamyfq','cig30use']


#These are continous columns
```

```python
cat_var = []
for i in df.columns:
    if i not in cont_var:
        cat_var.append(i)
cat_var

#these are categorical columns
```

### CHISQUARE

```python
import scipy.stats as stats
```

```python
pval = []
for i in cat_var:
    obs = pd.crosstab(df1['addprev'],df1[i])
    stat,p,dof,expected =stats.chi2_contingency(obs)
    pval.append(p)
    print(p)
```

```python
chisquare_table =pd.DataFrame()
chisquare_table['cat_independent_features']=cat_var
chisquare_table['p-value']=pval
chisquare_table
```

```python
df1.shape
```

```python
p_val = []
for i in cont_var:
    s1 = df1[df1['addprev']==0.0][i]
    s2=df1[df1['addprev']==1.0][i]
    test_stat,pval=stats.ttest_ind(s1,s2)
    p_val.append(pval)
    print(i,":",pval)
```

```python
con_ttest=pd.DataFrame()
con_ttest['Features'] = cont_var
con_ttest['p-value'] = p_val
con_ttest
```

```python
def model_build(model,X_train,X_test,y_train,y_test):
    model_=model.fit(X_train,y_train)
    y_pred_train=model_.predict(X_train)
    y_pred_test=model_.predict(X_test)
    y_pred_proba_train=model_.predict_proba(X_train)
    y_pred_proba_test=model_.predict_proba(X_test)
    print("Results on Train data: \n")
    print("Accuracy on train data :",accuracy_score(y_train, y_pred_train))
    print("\n Confusion matrix : \n",confusion_matrix(y_train,y_pred_train))
    print("\n Classification report : \n",classification_report(y_train,y_pred_train))
    print("ROC_AUC score :",roc_auc_score(y_train,y_pred_proba_train[:,1]))
    print("---------------------------")
    print("Results on Test data: \n")
    print("Accuracy on test data :",accuracy_score(y_test, y_pred_test))
    print("\n Confusion matrix : \n",confusion_matrix(y_test,y_pred_test))
    print("\n Classification report : \n",classification_report(y_test,y_pred_test))
    print("ROC_AUC score :",roc_auc_score(y_test,y_pred_proba_test[:,1]))
    cm=confusion_matrix(y_test,y_pred_test)
    print("sensitivity of algorithm on test data",round(cm[1,1]/float(cm[1,1]+cm[1,0]),3))
    print("specificty of algorithm on test data",round(cm[0,0]/float(cm[0,0]+cm[0,1]),3))
```
